```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
#Load Data
data = pd.read_csv("insurance.csv")
data.head()
```

```python
#Basic Info
print(data.info())
print(data.describe())
data.isna().sum()
```

```python
#Histograms (Continuous Variables
continuous_cols = ['age', 'bmi', 'charges']

plt.figure(figsize=(15,4))

for i, col in enumerate(continuous_cols):
    plt.subplot(1,3,i+1)
    sns.histplot(data[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)

plt.tight_layout()
plt.savefig("hist_continuous.png", dpi=300, bbox_inches='tight')
plt.show()

```

```python
#Boxplots (Outlier Detection)
plt.figure(figsize=(15,4))

for i, col in enumerate(continuous_cols):
    plt.subplot(1,3,i+1)
    sns.boxplot(x=data[col])
    plt.title(f'Boxplot of {col}')

plt.tight_layout()
plt.savefig("boxplots_continuous.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
#Countplots (Categorical Variables)
categorical_cols = ['sex', 'smoker', 'region']

plt.figure(figsize=(15,4))

for i, col in enumerate(categorical_cols):
    plt.subplot(1,3,i+1)
    sns.countplot(x=data[col])
    plt.title(f'Count of {col}')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.savefig("countplots_categorical.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
#Children Distribution (Discrete Numerical)
plt.figure(figsize=(6,4))
sns.countplot(x=data['children'])
plt.title("Distribution of Children")
plt.tight_layout()
plt.savefig("children_distribution.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
#Scatterplots (Continuous vs Target)
plt.figure(figsize=(15,4))

for i, col in enumerate(['age','bmi']):
    plt.subplot(1,2,i+1)
    sns.scatterplot(x=data[col], y=data['charges'])
    plt.title(f'{col} vs Charges')

plt.tight_layout()
plt.savefig("scatter_continuous_vs_target.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
#Boxplots (Categorical vs Target)
plt.figure(figsize=(15,4))

for i, col in enumerate(['sex','smoker','region']):
    plt.subplot(1,3,i+1)
    sns.boxplot(x=data[col], y=data['charges'])
    plt.title(f'{col} vs Charges')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.savefig("categorical_vs_target.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
#Correlation Heatmap 
numeric_data = data.select_dtypes(include='number')

plt.figure(figsize=(8,6))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')

plt.title("Correlation Heatmap")
plt.tight_layout()
plt.savefig("correlation_heatmap.png", dpi=300, bbox_inches='tight')
plt.show()
```
